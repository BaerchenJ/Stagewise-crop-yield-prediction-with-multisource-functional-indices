```python
## write a loop
from pandas import read_csv

import csv

import numpy as np

Newlist = []

df = read_csv('.../Illisoy/CPC-SOY/CondiResults/2019Condi.csv')

d = df.values

y = [d[d[:,0]==k] for k in np.unique(d[:,0])]

for coarray in y:
    
    y0 = coarray[coarray[:, 1].argsort()]
    
    # Removing duplicate rows
    new = [tuple(row) for row in y0]
    
    res = np.unique(new,axis=0)
    
    res1 = res.tolist()

    Newlist.append(res1)
    
print(Newlist)
```

```python
csv_filename = '.../Illisoy/CPC-SOY/CondiResults/NewCondi20191.csv'
fieldnames = ['COUNTY','WEEK','INDEX']
with open(csv_filename, mode='w') as csv_file:
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    writer.writeheader()
    for countylist in Newlist:
        for i in countylist:
            writer.writerow({fieldnames[0]: i[0], fieldnames[1]: i[1], fieldnames[2]: i[2]}) 
```

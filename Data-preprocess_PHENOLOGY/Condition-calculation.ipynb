```python
Daterate = []

for countyname in CSlist:
    
    directory = '.../Illisoy/Illinois_shp/county_shape_files/ToJSON'

    mydir = '.../Illisoy/PHENOLOGY/2022Con/County'
        
    IndiCountylist = []
    
    def substring_before(s, delim):
        return s.partition(delim)[0]

    IndiCountylist += [each for each in os.listdir(mydir) if substring_before(Path(each).stem, 'soy') == Path(os.path.join(directory, countyname)).stem
        and each.endswith('.tif')]
        
    for file in IndiCountylist:
        
        with rio.open(os.path.join(mydir, file)) as eachdate:
    
            EDarray = eachdate.read(1)
    
            EDarray[EDarray==-9999] = np.NaN
            
            if  math.isnan(np.nanmean(EDarray)):
                print(f"{Path(file).stem} EDmean == NaN.") 
            else:
                EDmean = np.nanmean(EDarray)
            
                EDMvec = {'County': Path(countyname).stem, 'Week': Path(file).stem[-2:],
                          'EDmean': EDmean}   
                
            Daterate.append(EDMvec)
            
print(Daterate)
```

```python
EDMvec = {'County': Path(countyname).stem, 'Week': Path(file).stem[-2:],
          'EDmean': EDmean}

# save as csv.
with open('.../Illisoy/PHENOLOGY/CondiResults/2022CondiLIV.csv', 'w') as f:  
    w = csv.DictWriter(f, EDMvec.keys())
    w.writeheader()
    w.writerows(Daterate)
```

```python
oriwitable = pd.read_excel("wi.xlsx")
```

```python
CRarr = np.array("weekCR")
GDDarr = np.array("weekGDD")
SRarr = np.array("weekSR")
VParr = np.array("weekVP")
```

```python
for nr in range(0, len(normaldf)):
    carray = np.array(normaldf.iloc[nr, :][3:29])
    CRarr = np.hstack((CRarr, carray))
    garray = np.array(normaldf.iloc[nr, :][29:55])
    GDDarr = np.hstack((GDDarr, garray))
    sarray = np.array(normaldf.iloc[nr, :][55:81])
    SRarr = np.hstack((SRarr, sarray))
    varray = np.array(normaldf.iloc[nr, :][81:107])
    VParr = np.hstack((VParr, varray))
```

```python
oriwitable[['CR']] = pd.DataFrame(CRarr[1:(len(CRarr))])
oriwitable[['GDD']] = pd.DataFrame(GDDarr[1:(len(GDDarr))])
oriwitable[['SR']] = pd.DataFrame(SRarr[1:(len(SRarr))])
oriwitable[['VP']] = pd.DataFrame(VParr[1:(len(VParr))])
```

```python
oriwitable.to_excel('nowi.xlsx', index=False)
```
